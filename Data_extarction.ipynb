```python
import os
import zipfile

# Specify the directory containing the NASDAQ 100 stock files
nasdaq_folder_path = 'nasdaq'

# List of NASDAQ 100 stock tickers
nasdaq_100_stocks = [
    "AAPL", "MSFT", "AMZN", "GOOGL", "FB", "TSLA", "NVDA", "PYPL", "ASML", 
    "ADBE", "CMCSA", "CSCO", "INTC", "PEP", "AVGO", "TMUS", "NFLX", "CHTR", "TXN", 
    "ACN", "ADP", "QCOM", "COST", "AMGN", "INTU", "AMAT", "AMD", "MU", "LRCX", 
    "FISV", "SBUX", "GILD", "MDLZ", "BKNG", "ADI", "ISRG", "VRTX", "REGN", "ATVI", 
    "CSX", "MRNA", "ADSK", "ILMN", "EXC", "ZM", "MELI", "KHC", "WBA", "EBAY", 
    "ORLY", "IDXX", "LULU", "EA", "SNPS", "WDAY", "KLAC", "CTSH", "ROST", "MNST", 
    "MAR", "PAYX", "DXCM", "XEL", "NTES", "DOCU", "SPLK", "ANSS", "FAST", "CDNS", 
    "NXPI", "XRAY", "PCAR", "VRSK", "CDW", "TEAM", "BMRN", "SGEN", "SIRI", "DLTR", 
    "ALXN", "INCY", "TCOM", "ULTA", "FOX", "CHKP", "MXIM", "CPRT", "CERN", "EXPE", 
    "VRSN", "TTWO", "TCBK", "LBTYK", "UAL", "FOX", "LBTYA", "CTAS", "META"
]

# Create a directory to store extracted files
extracted_dir = 'extracted_nasdaq_files'
os.makedirs(extracted_dir, exist_ok=True)

# Extract CSV files for NASDAQ 100 stocks
for stock_ticker in nasdaq_100_stocks:
    # Construct the filename
    file_name = f'{stock_ticker}.csv'
    
    # Check if the file exists in the folder
    file_path = os.path.join(nasdaq_folder_path, file_name)
    if os.path.isfile(file_path):
        # Copy the file to the extraction directory
        target_file_path = os.path.join(extracted_dir, file_name)
        with open(file_path, 'rb') as source_file, open(target_file_path, 'wb') as target_file:
            target_file.write(source_file.read())
        print(f'Extracted {file_name}')
    else:
        print(f'File {file_name} not found')

# Create a new zip archive with extracted files
output_zip_path = 'op-nasdaq_100.zip'
with zipfile.ZipFile(output_zip_path, 'w') as zip_ref:
    # Add extracted files to the new zip archive
    for root, dirs, files in os.walk(extracted_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zip_ref.write(file_path, os.path.relpath(file_path, extracted_dir))

print(f'Extracted NASDAQ 100 stock files are stored in {output_zip_path}')
```
